```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
import seaborn as sns
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
sns.countplot(x='Sex', data=titanic)
```

```python
sns.barplot(x='Sex', y='Survived', data=titanic)
```

```python
sns.boxplot(x='Sex', y='Age', data=titanic,palette='rainbow')
```

```python
sns.boxplot(data=titanic,palette='rainbow',orient='h')
```

```python
sns.stripplot(x="Pclass",y="Fare",data=titanic)
```

```python
sns.swarmplot(x="Pclass",y="Fare",data=titanic)
```

```python
sns.swarmplot(x="Pclass",y="Fare",hue="Sex",data=titanic)
```

```python
#heat map  
#it is 2 d representation of data
# here value is represented using colors
#it helps us determine how two variables are corelated with each other
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.heatmap(iris.corr())
```

```python
sns.heatmap(iris.corr(),cmap='coolwarm',annot=True)
```

```python
#to understand distribution of dataset
# we use distplot
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips.total_bill)
```

```python
sns.distplot(tips.total_bill,kde=False,bins=30)
```

```python
# to see distribution of whole dataset
# we use pairplot
# it gives distribution in each column
```

```python
sns.pairplot(tips, hue='sex', palette='husl')
```
